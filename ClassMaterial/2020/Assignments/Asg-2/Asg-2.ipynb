# Here is how to write a test

```python
# tests
tests = [ nthnumeric(i, Sigma={'0','1'}) for i in range(17)]
# testdfa
testdfa = md2mc('''DFA 
I : 0 -> F
F : 0 -> I
I : 1 -> I
F : 1 -> F''')

dotObj_dfa(testdfa)
```

```python
for s in tests:
    if accepts_dfa(testdfa, s):
        print("testdfa accepts ", s)
    else:
        print("testdfa does not accept ", s)
```
